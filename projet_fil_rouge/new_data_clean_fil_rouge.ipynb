### Sauvegarde des données en CSV avant intégration en Base

```python
for (repertoire, sousRepertoires, fichiers) in walk(destination_path):
    for file in fichiers:
        print(file)
        index = file.find('jwl')
        if index != -1:
            no = file.split("_jwl")[0]
        else:
            no = file.split("_dwl")[0]
        data = "data_"+str(no)
        data = lasio.read(str(destination_path)+'/'+str(file))
        
        if data.well.WELL.value != '':
            well_1 = data.well.WELL.value.strip().replace(' ', '').split('(')[0]
            well = well_1.replace('UK_', '')
        else:
            well_1 = no.replace('-_', '-')
            well   = well_1.replace('_', '/')
        
        #data.to_csv("output/out_"+str(no)+'.csv', units_loc='[]')
        data.to_csv("output/out_"+str(no)+'.csv')
        data_csv = "data_csv_"+str(no)
        data_csv = pd.read_csv("output/out_"+str(no)+'.csv')
        
        if 'BOREHOLE-DEPTH' in data_csv.columns:
            data_csv = data_csv.rename(columns={'BOREHOLE-DEPTH': 'DEPTH'})
        
        if 'DEPT' in data_csv.columns:
            data_csv = data_csv.rename(columns={'DEPT': 'DEPTH'})
        
        if 'GR' not in data_csv.columns:
            for gr in alias['GR']:
                if str(gr) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(gr): 'GR'})
                    break
                    
        if 'CAL' not in data_csv.columns:
            for cal in alias['CAL']:
                if str(cal) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(cal): 'CAL'})
                    break
                    
        if 'DCOR' not in data_csv.columns:
            for dcor in alias['DCOR']:
                if str(dcor) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(dcor): 'DCOR'})
                    break
        
        if 'DENS' not in data_csv.columns:
            for dens in alias['DENS']:
                if str(dens) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(dens): 'DENS'})
                    break
                    
        if 'DTC' not in data_csv.columns:
            for dtc in alias['DTC']:
                if str(dtc) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(dtc): 'DTC'})
                    break
                    
        if 'DTS' not in data_csv.columns:
            for dts in alias['DTS']:
                if str(dts) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(dts): 'DTS'})
                    break
                    
        if 'PE' not in data_csv.columns:
            for pe in alias['PE']:
                if str(pe) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(pe): 'PE'})
                    break
                    
        if 'PHIN' not in data_csv.columns:
            for phin in alias['PHIN']:
                if str(phin) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(phin): 'PHIN'})
                    break
        
        if 'RESD' not in data_csv.columns:
            for resd in alias['RESD']:
                if str(resd) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(resd): 'RESD'})
                    break
                    
        if 'RESM' not in data_csv.columns:
            for resm in alias['RESM']:
                if str(resm) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(resm): 'RESM'})
                    break
                    
        if 'RESS' not in data_csv.columns:
            for ress in alias['RESS']:
                if str(ress) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(ress): 'RESS'})
                    break
                    
        if 'RESMIC' not in data_csv.columns:
            for resmic in alias['RESMIC']:
                if str(resmic) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(resmic): 'RESMIC'})
                    break
                    
        if 'SP' not in data_csv.columns:
            for sp in alias['SP']:
                if str(sp) in data_csv.columns:
                    data_csv = data_csv.rename(columns={str(sp): 'SP'})
                    break
                    
        #cols = ['DEPTH [ft]','GR [gAPI]', 'NPHI [LPU]', 'RHOB [g/cc]', 'SP [mV]']
        data_use = pd.DataFrame(data_csv, columns = ['DEPTH'])
        data_use['DEPTH']=  pd.to_numeric(data_csv['DEPTH'],  errors='coerce')

        
        if 'GR' in data_csv.columns:
            data_use['GR'] = data_csv['GR']
            data_use['GR']=  pd.to_numeric(data_use['GR'], errors='coerce')
        else:
            data_use['GR']=  -9999
            
        
        if 'CAL' in data_csv.columns:
            data_use['CAL'] = data_csv['CAL']
            data_use['CAL']=  pd.to_numeric(data_use['CAL'], errors='coerce')
        else:
            data_use['CAL']=  -9999

            
        if 'DCOR' in data_csv.columns:
            data_use['DCOR']= data_csv['DCOR']
            data_use['DCOR']=  pd.to_numeric(data_use['DCOR'], errors='coerce')
        else:
            data_use['DCOR']=  -9999
            
                
        if 'DENS' in data_csv.columns:
            data_use['DENS']= data_csv['DENS']
            data_use['DENS']=  pd.to_numeric(data_use['DENS'], errors='coerce')
        else:
            data_use['DENS']=  -9999
            
                
        if 'DTC' in data_csv.columns:
            data_use['DTC']= data_csv['DTC']
            data_use['DTC']=  pd.to_numeric(data_use['DTC'], errors='coerce')
        else:
            data_use['DTC']=  -9999
            
        
        if 'DTS' in data_csv.columns:
            data_use['DTS']= data_csv['DTS']
            data_use['DTS']=  pd.to_numeric( data_use['DTS'], errors='coerce')
        else:
            data_use['DTS']=  -9999
            
        
        if 'PE' in data_csv.columns:
            data_use['PE'] =  data_csv['PE']
            data_use['PE']=  pd.to_numeric(data_use['PE'], errors='coerce')
        else:
            data_use['PE']=  -9999
                   
                
        if 'PHIN' in data_csv.columns:
            data_use['PHIN'] =  data_csv['PHIN']
            data_use['PHIN']=  pd.to_numeric(data_use['PHIN'], errors='coerce')
        else:
            data_use['PHIN']=  -9999
                 
        
        if 'RESD' in data_csv.columns:
            data_use['RESD'] =  data_csv['RESD']
            data_use['RESD']=  pd.to_numeric(data_use['RESD'], errors='coerce')
        else:
            data_use['RESD']=  -9999
            
         
        if 'RESM' in data_csv.columns:
            data_use['RESM'] =  data_csv['RESM']
            data_use['RESM']=  pd.to_numeric(data_use['RESM'], errors='coerce')
        else:
            data_use['RESM']=  -9999
        

        if 'RESS' in data_csv.columns:
            data_use['RESS'] =  data_csv['RESS']
            data_use['RESS']=  pd.to_numeric(data_use['RESS'], errors='coerce')
        else:
            data_use['RESS']=  -9999
                

        if 'RESMIC' in data_csv.columns:
            data_use['RESMIC'] =  data_csv['RESMIC']
            data_use['RESMIC']=  pd.to_numeric(data_use['RESMIC'], errors='coerce')
        else:
            data_use['RESMIC']=  -9999
                     

        if 'SP' in data_csv.columns:
            data_use['SP'] =  data_csv['SP']
            data_use['SP']=  pd.to_numeric(data_use['SP'], errors='coerce')
        else:
            data_use['SP']=  -9999
        
        ## Transformation des profondeurs en mètre
        if well not in ['A1', 'A2', 'A3', 'A4','A6', 'A9','N1', 'N2', 'N3', 'N9', 'N14', 'N18']:
            data_use['depth_m'] = data_use['DEPTH'] * 0.3048
        else:
            data_use['depth_m'] = data_use['DEPTH']
        
        ## ajout champ well
        data_use['well'] = well
        data_use['litho'] = 0
        data_use.to_csv("output_clean/out_clean"+str(no)+'.csv')
```
